```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
drive = pd.read_csv("drive.csv")
```

```python
drive
```

```python
plt.scatter(drive["speed"], drive["risk"], color = "red" )
plt.xlabel("Speed")
plt.ylabel("Risk")
plt.title("Risk Analysis")
```

```python
speed = drive[["speed"]]
risk = drive[["risk"]]
```

```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
speedtrain, speedtest, risktrain, risktest = train_test_split(speed, risk, test_size=.40, random_state=1)
```

```python
speedtrain
```

```python
speedtest
```

```python
risktrain
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regra = LinearRegression()
```

```python
regra.fit(speedtrain, risktrain)
```

```python
regra.predict(speedtest)
```

```python
speedtest
```

```python
risktest
```

```python
plt.plot(drive.speed, regra.predict(drive[["speed"]]))
plt.scatter(drive["speed"], drive["risk"], color = "red")
plt.xlabel("Speed")
plt.ylabel("Risk")
plt.title("Risk Analysis")
```
